```python
import pandas as pd
```

```python
df=pd.read_csv('/content/drive/MyDrive/swords_v2/stopWords_v2.csv')
```

```python
data={'index':[],
      'data':[],
      'r1':[],
      'r2':[],
      'r3':[],
      'r4':[]    
}
```

```python
import nltk
nltk.download('wordnet')
nltk.download('omw-1.4')
from nltk.stem import WordNetLemmatizer 
lemma=WordNetLemmatizer()
```

```python
for i in range(20):
  str=df.iloc[i][1]
  str1=""

  wd=""
  for word in str.split():
    p=""
    if word.endswith('.'):
      p=p+"."
      wd=word.replace(".","")
    elif word.endswith(','):
      p=p+","
      wd=word.replace(",","")
    elif word.endswith('?'):
      p=p+"?"
      wd=word.replace("?","")
    elif word.endswith('!'):
      p=p+"!"
      wd=word.replace("!","")
    elif word.endswith(')'):
      p=p+")"
      wd=word.replace(")","")
    else:
      p=""
      wd=word
    

    w=lemma.lemmatize(wd)
    str1=str1+w+p+" "
  print(str)
  print("\n\n")
  print(str1)
  data['index'].append(df.iloc[i][0])
  data['data'].append(str1)
  data['r1'].append(df.iloc[i][2])
  data['r2'].append(df.iloc[i][3])
  data['r3'].append(df.iloc[i][4])
  data['r4'].append(df.iloc[i][5])
```

```python
for i in range(20,50):
  str=df.iloc[i][1]
  str1=""

  wd=""
  for word in str.split():
    p=""
    if word.endswith('.'):
      p=p+"."
      wd=word.replace(".","")
    elif word.endswith(','):
      p=p+","
      wd=word.replace(",","")
    elif word.endswith('?'):
      p=p+"?"
      wd=word.replace("?","")
    elif word.endswith('!'):
      p=p+"!"
      wd=word.replace("!","")
    elif word.endswith(')'):
      p=p+")"
      wd=word.replace(")","")
    else:
      p=""
      wd=word
    

    w=lemma.lemmatize(wd)
    str1=str1+w+p+" "
  print(str)
  print("\n\n")
  print(str1)
  data['index'].append(df.iloc[i][0])
  data['data'].append(str1)
  data['r1'].append(df.iloc[i][2])
  data['r2'].append(df.iloc[i][3])
  data['r3'].append(df.iloc[i][4])
  data['r4'].append(df.iloc[i][5])
```

```python
d = pd.DataFrame.from_dict(data)
d.set_index('index',inplace = True)
d.sort_index(ascending=True, inplace=True)
```
