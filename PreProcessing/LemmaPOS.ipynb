```python
from nltk.stem import WordNetLemmatizer
lemma = WordNetLemmatizer()
```

```python
import pandas as pd
```

```python
df=pd.read_csv('/content/drive/MyDrive/(DUC 2002 A)Input Files/NSW.csv')
```

```python
data={'index':[],
      'file':[],
      'doc':[],
      'doc_dump':[],
      '1_10':[],
      '1_50':[],
      '1_100':[],
      '1_200':[],
      '2_10':[],
      '2_50':[],
      '2_100':[], 
      '2_200':[],
      '3_10':[],
      '3_50':[],
      '3_100':[],
      '3_200':[]       
}
```

```python
from nltk.corpus.reader.wordnet import VERB, NOUN, ADJ, ADV
dict_pos_map = {
    
    'NN' : 'n',
    'NN' : 'n',
    'VB' : 'v', 
    'VBG' : 'v',
    'VBN' : 'v',
    'VBP' : 'v',
    'VBZ' : 'v',
    'VBD': 'v',
    'JJ' : 'a',
    'JJR' : 'a',
    'JJS' : 'a',

    'RB': ADV
}
```

```python
for i in range(20):
  str=df.iloc[i][2]
  l1 = str.split()
  tl1 = nltk.pos_tag(l1)
  print(tl1)
  l2 = []
  #i = 0

  normalized_tokens = []
  for each_seq in tl1:
    
    temp = each_seq[0]
    if each_seq[1] == "NNP" or each_seq[1] == "NNPS" or each_seq[1] not in dict_pos_map.keys():
      temp = each_seq[0]
    else:
      temp = lemma.lemmatize(each_seq[0].lower(), pos = dict_pos_map[each_seq[1]])
        
    normalized_tokens.append(temp)
  l2.append(normalized_tokens)
  optstr = ""
  for x in l2:
    for y in x:
      optstr = optstr + y + " "

  data['index'].append(df.iloc[i][0])
  data['file'].append(df.iloc[i][1])
  data['doc'].append(optstr)
  data['doc_dump'].append(df.iloc[i][3])
  data['1_10'].append(df.iloc[i][4])
  data['1_50'].append(df.iloc[i][5])
  data['1_100'].append(df.iloc[i][6])
  data['1_200'].append(df.iloc[i][7])
  data['2_10'].append(df.iloc[i][8])
  data['2_50'].append(df.iloc[i][9])
  data['2_100'].append(df.iloc[i][10])
  data['2_200'].append(df.iloc[i][11])
  data['3_10'].append(df.iloc[i][12])
  data['3_50'].append(df.iloc[i][13])
  data['3_100'].append(df.iloc[i][14])
  data['3_200'].append(df.iloc[i][15])

  print("\n\n")
```
