```python
import logging
import pandas as pd
from simpletransformers.seq2seq import Seq2SeqModel, Seq2SeqArgs
```

```python

model_args = Seq2SeqArgs()
model_args.num_train_epochs = 10
model_args.no_save = True
model_args.evaluate_generated_text = True
model_args.evaluate_during_training = True
model_args.evaluate_during_training_verbose = True
model_args.min_length=90
model_args.max_length=120


model = Seq2SeqModel(
    encoder_decoder_type="bart",
    encoder_decoder_name="facebook/bart-large",
    args=model_args,
    use_cuda=False,
)
```

```python
training_data=[]
```

```python
for i in range(0,46):
  if not pd.isna(df.iloc[i][10]):
    df.iloc[i][2] = "\"" + df.iloc[i][2] + "\""
    df.iloc[i][10] = "\"" + df.iloc[i][10] +  "\""
    training_data.append([df.iloc[i][2],df.iloc[i][10]])

  
```

```python
train_df = pd.DataFrame(
    training_data, columns=["input_text", "target_text"]
)
```

```python
eval_data =[]
```

```python
for i in range(46,58):
  if not pd.isna(df.iloc[i][10]):
    df.iloc[i][2] = "\"" + df.iloc[i][2] + "\""
    df.iloc[i][10] = "\"" + df.iloc[i][10] + "\""
    eval_data.append([df.iloc[i][2],df.iloc[i][10]])
```

```python
eval_df = pd.DataFrame(eval_data, columns=["input_text", "target_text"])
```

```python
model.train_model(train_df, eval_data=eval_df)
```

```python
results = model.eval_model(eval_df)
```
