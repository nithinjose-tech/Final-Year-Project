```python
from nltk.stem import WordNetLemmatizer
lemma = WordNetLemmatizer()
```

```python
import pandas as pd
```

```python
df=pd.read_csv('/content/drive/MyDrive/swords_v2/stopWords_v2.csv')
```

```python
data={'index':[],
      'data':[],
      'r1':[],
      'r2':[],
      'r3':[],
      'r4':[]   
}
```

```python
from nltk.corpus.reader.wordnet import VERB, NOUN, ADJ, ADV
dict_pos_map = {
    
    'NN' : 'n',
    'NN' : 'n',
    'VB' : 'v', 
    'VBG' : 'v',
    'VBN' : 'v',
    'VBP' : 'v',
    'VBZ' : 'v',
    'VBD': 'v',
    'JJ' : 'a',
    'JJR' : 'a',
    'JJS' : 'a',

    'RB': ADV
}
```

```python
for i in range(20):
  str=df.iloc[i][1]
  l1 = str.split()
  tl1 = nltk.pos_tag(l1)
  print(tl1)
  l2 = []
  #i = 0

  normalized_tokens = []
  for each_seq in tl1:
    
    temp = each_seq[0]
    if each_seq[1] == "NNP" or each_seq[1] == "NNPS" or each_seq[1] not in dict_pos_map.keys():
      temp = each_seq[0]
    else:
      temp = lemma.lemmatize(each_seq[0].lower(), pos = dict_pos_map[each_seq[1]])
        
    normalized_tokens.append(temp)
  l2.append(normalized_tokens)
  optstr = ""
  for x in l2:
    for y in x:
      optstr = optstr + y + " "

  print(l1)
  print(l2)
  print(optstr)
  print(df.iloc[i][0])
  a = df.iloc[i][0]
  data['index'].append(df.iloc[i][0])
  data['data'].append(optstr)
  data['r1'].append(df.iloc[i][2])
  data['r2'].append(df.iloc[i][3])
  data['r3'].append(df.iloc[i][4])
  data['r4'].append(df.iloc[i][5])
  
  print("\n\n")
```
