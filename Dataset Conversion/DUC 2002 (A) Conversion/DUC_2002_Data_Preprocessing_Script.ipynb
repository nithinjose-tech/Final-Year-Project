```python
root_dir = "/content/drive/MyDrive/DUC 2002/"
```

```python
docs_dir = root_dir + "docs/"
summary_dir = root_dir + "summaries/"
```

```python
from pathlib import Path
from bs4 import BeautifulSoup
import pandas as pd


h = {}

for i in ['file', 'doc', 'doc_dump', '1_200', '1_100', '1_50', '1_10', '2_10', '2_100', '2_50', '2_200', '3_200', '3_50', '3_100', '3_10']:
  h[i] = []
 
for path in Path(docs_dir).iterdir():
    # print(path.name)
    if path.is_dir():
        res = []
        file_res = []
        for file_path in Path(path).iterdir():
          with open(file_path) as f:
            file_res.append(file_path)
            soup = BeautifulSoup(f.read(), 'html.parser')
            res.append(soup.find_all('text')[0].string)

        d = 1
        
        h['file'].append(path.name)
        h['doc'].append("\n".join([x.strip() for x in res if x is not None]))
        h['doc_dump'].append(",".join([str(x.name) for x in file_res]))
        
        for sum_path in Path(summary_dir).iterdir():
          if sum_path.name.startswith(path.name):
            for sum_path_sub in Path(sum_path).iterdir():
              if sum_path_sub.name.startswith("per"): continue
              if sum_path_sub.name.endswith("e"): continue
              with open(sum_path_sub) as f:
                soup = BeautifulSoup(f.read(), 'html.parser').sum.string

                h[f"{d}_{sum_path_sub.name}"].append(soup.strip())

            d += 1
        
        max_val = len(h['file'])
        for x in h:
          if(len(h[x]) != max_val):
            h[x].append(None)

print([len(h[x]) for x in h])
print([x for x in h])
```

```python
df = pd.DataFrame.from_dict(h)
```
